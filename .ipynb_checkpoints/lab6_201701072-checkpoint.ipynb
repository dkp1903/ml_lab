```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
di = h5py.File('faceimages.mat','r')
```

```python
#we need to convert it into the list to view it
list(di.keys())
```

```python
#data contains information about images
x = np.array(di["data"][:])
```

```python
print(di['width'][0],di['height'][0])
```

```python
print(di['data'].shape)
```

```python
#deleting last row because 122*92 = 10304
x = np.delete(x,10304,0)
#we have features row wise and different different images column wise
```

```python
y = x.reshape((92,112,400))
first = y[:,:,0]
plt.imshow(first.T,cmap='gray')
plt.show()
```

```python
#alternative is you use singular value decomposition
x_cov = np.cov(x)
print(x_cov.shape)
```

```python
eigvals, eigvecs = np.linalg.eigh(x_cov)
print(eigvecs.shape)
```

```python
index = eigvals.argsort()[::-1]
eigvals = eigvals[index]
eigvecs = eigvecs[:,index]
```

```python
U = eigvecs
Y = U.T@x
print(Y.shape)
```

```python
#taking first 50 eigen value and constructing the image 
recX50 = U[:,:50]@Y[:50,0]
compres = recX50.reshape((92,112))
plt.imshow(compres.T,cmap='gray')
plt.show()
print(np.sqrt(np.sum(abs(recX50-x[:,0])**2)/len(recX50)))

#taking first 100
recX100 = U[:,:100]@Y[:100,0]
compres = recX100.reshape((92,112))
plt.imshow(compres.T,cmap='gray')
plt.show()
print(np.sqrt(np.sum(abs(recX100-x[:,0])**2)/len(recX100)))

#taking first 400+
recX400plus = U[:,:10304]@Y[:10304,0]
compres = recX400plus.reshape((92,112))
plt.imshow(compres.T,cmap='gray')
plt.show()
print(np.sqrt(np.sum(abs(recX400plus-x[:,0])**2)/len(recX400)))
```
