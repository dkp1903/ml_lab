```python
import pandas as pd
import io
df2=pd.read_excel('/home/rohan/Downloads/datafPr1.xlsx')
x=df2['x'].to_numpy()
x=(x-np.mean(x))/np.std(x)
y=df2['y'].to_numpy()
theta0=np.arange(-30,30,0.1)
theta1=np.arange(-30,30,0.1)
theta0,theta1=np.meshgrid(theta0,theta1)
m=len(x)
ltheta=(y[0]-(theta0+theta1*(x[0])))**2
for i in range(1,m):
    ltheta=ltheta+(y[i]-(theta0+theta1*(x[i])))**2
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(theta1,theta0,ltheta, 500, cmap='binary')
plt.show()
```

```python
A,B = np.unravel_index(np.argmin(ltheta, axis=None), ltheta.shape)
print(theta0[A][B])
print(theta1[A][B])
```

```python
temp = np.ones((len(x),1),dtype='int')
X = np.c_[temp,x]
theta = np.linalg.inv((X.T)@X)@(X.T)@y
print(theta)
print(np.sum((y-(theta[0]+theta[1]*(x)))**2))
```
