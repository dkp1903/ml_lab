```python
import numpy as np
import matplotlib.pyplot as plt 
import math as mt
```

```python
k = 1000
x = np.linspace(0, (2*np.pi)/(1.5*1000)*k, k)
f = x + 2*np.sin(1.5*x)
plt.plot(x,f)
plt.show()
```

```python
r = np.ones((x.shape[0],1))
r.reshape(x.shape[0],1)
xx = np.c_[r,x]
```

```python
predictions = np.zeros((1000,k))
for i in range(1000):
    noise = np.random.normal(0,mt.sqrt(2),x.shape[0])
    y = f + noise
    thh = np.linalg.pinv((xx.T@xx))@xx.T@y
    predictions[i,:] = thh[0] + thh[1]*x
```

```python
avg_pred_y=np.mean(predictions,axis=0) 
plt.plot(x,f)
plt.plot(x,avg_pred_y)
plt.show()
```

```python
bias=np.mean(abs(f-avg_pred_y))
print(bias)
```

```python
var=0
for i in range(0,1000):
    var=var+np.mean((predictions[i,:]-avg_pred_y)**2)
var=var/1000
print(var)
```

```python
mse=0
for i in range(0,1000):
    mse=mse+np.mean((predictions[i,:]-f)**2)
mse=mse/1000
print(mse)
print(bias**2+var)
```

```python
bias=np.mean(abs(f-avg_pred_y))
print(bias)
```
